## Desafío 3

```python
count_div_2_o_3 = []
count_div_2_y_3 = []
count_div_3_not_2 = []
count_div_not_2_not_3 = []
for i in array2:
    if i % 2 == 0 or i % 3 == 0:
        count_div_2_o_3.append(i)
    if i % 2 == 0 and i % 3 == 0:
        count_div_2_y_3.append(i)
    if i % 3 == 0 and i % 2 != 0:
        count_div_3_not_2.append(i)
    if i % 2 != 0 and i % 3 != 0:
        count_div_not_2_not_3.append(i)

print(count_div_2_o_3)
print(count_div_2_y_3)
print(count_div_3_not_2)
print(count_div_not_2_not_3)
```

## Desafío 5

```python
import pandas as pd
df = pd.read_csv('flights.csv')
media = df['passengers'].mean()
print(media)
```

```python
df['underperforming'] = 0
for index, serie in df.iterrows():
    if serie['passengers'] < media:
        df.at[index, 'underperforming'] = 1
```

## Desafío 6

```python
media_general = df['passengers'].mean()
devstd_general = df['passengers'].std()
print(media_general, devstd_general)
```

```python
df['outlier'] = 0
for index, serie in df.iterrows():
    if serie['passengers'] < media_general - devstd_general or serie['passengers'] > media_general + devstd_general:
        df.at[index, 'outlier'] = 1
casos_extremos = df[df['outlier'] == 1]
print("La cantidad de casos extremos son {}".format(len(casos_extremos)))
```
